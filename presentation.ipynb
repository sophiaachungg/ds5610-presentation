### Data Cleaning & Wrangling

Before merging all 4 CSVs together, I dropped the columns that provide the lower and upper 95% uncertainty levels. Because I will be taking averages (and showing standard deviation in one of the analyses), the data lower and upper 95% uncertainty levels are a bit redundant and unnecessary for the overall goals of the analyses.

```python
# drop lower/upper columns
# in hypertension - drop columns age, lower 95% uncertainty and upper 95% uncertainty interval
# in blood pressure, cholesterol, diabetes - drop lower 95% uncertainty and upper 95% uncertainty interval
blood_pressure = blood_pressure.drop(columns = ['Lower 95% uncertainty interval',
                                                'Upper 95% uncertainty interval'])
cholesterol = cholesterol.drop(columns = ['Mean total cholesterol lower 95% uncertainty interval (mmol/L)',
                                          'Mean total cholesterol upper 95% uncertainty interval (mmol/L)',
                                          'Mean non-HDL cholesterol lower 95% uncertainty interval (mmol/L)',
                                          'Mean non-HDL cholesterol upper 95% uncertainty interval (mmol/L)',
                                          'Mean HDL cholesterol lower 95% uncertainty interval (mmol/L)',
                                          'Mean HDL cholesterol upper 95% uncertainty interval (mmol/L)'])
diabetes = diabetes.drop(columns = ['Lower 95% uncertainty interval',
                                    'Upper 95% uncertainty interval'])
hypertension = hypertension.drop(columns = ['Age', 
                                            'Prevalence of hypertension lower 95% uncertainty interval', 
                                            'Prevalence of hypertension upper 95% uncertainty interval',
                                            'Proportion of diagnosed hypertension among all hypertension lower 95% uncertainty interval',
                                            'Proportion of diagnosed hypertension among all hypertension upper 95% uncertainty interval',
                                            'Proportion of treated hypertension among all hypertension lower 95% uncertainty interval',
                                            'Proportion of treated hypertension among all hypertension upper 95% uncertainty interval',
                                            'Proportion of controlled hypertension among all hypertension lower 95% uncertainty interval',
                                            'Proportion of controlled hypertension among all hypertension upper 95% uncertainty interval',
                                            'Proportion of untreated stage 2 hypertension among all hypertension lower 95% uncertainty interval',
                                            'Proportion of untreated stage 2 hypertension among all hypertension upper 95% uncertainty interval'])
```

```python
# make a new column that functions like the ID for the row, makes merging easier later on
# this ID is made up of the ISO initials + _ + year
# e.g., AFG_1990 for row index 1 of hypertension table
dfs = [blood_pressure, cholesterol, diabetes, hypertension]
for df in dfs:
    for row in range(len(df)):
        df.loc[row, 'ID'] = df.loc[row, 'ISO'] + '_' + str(df.loc[row, 'Year']) + '_' + df.loc[row, 'Sex']
    df = df.drop(columns = ['ISO', 'Year', 'Sex', 'Country/Region/World'], inplace=True) 
```

```python
# merge all 4 tables together on the new identifier column
merged_df = pd.merge(blood_pressure, cholesterol, on='ID', how='outer')
merged_df.head(5)
```

```python
merged_df = pd.merge(merged_df, diabetes, on='ID', how='outer')
merged_df.head(5)
```

```python
merged_df = pd.merge(merged_df, hypertension, on='ID', how='outer')
merged_df.head(5)
```

Originally, I normalized the data prior to dealing with missing values. This was due to the fact that the columns relating to prevalence/proportion of the populations ranged from 0 to 1, while the mean cholesterol values in mmol/L did not have the same range. Considering the visualizations I decided to use in the end, I decided to ultimately keep the data as is for the cholesterol columns because the normalized values did not make sense, and I was not intending on mixing an analysis of cholesterol levels and prevalence/proportion columns.

```python
# dealing with missing data
df = merged_df.dropna()
df = df.reset_index(drop=True)
df.head(5)
```

Originally, I used KNN Imputation to impute missing values. When conducting the exploratory data analysis portion and visualizing the imputed data, though, I noticed that the imputed values did not make sense in relation to the overall trend of the available data. I decided to drop the rows with missing values and re-analyze the data. Removing the missing values did not significantly impact the visualizations I had, but improved clarity on the time-series analysis.

```python
# break up ID column into categorical columns
df[['Country','Year','Sex']] = df['ID'].str.split('_', expand = True)
df = df.drop('ID', axis = 1)
new_order = ['Country', 'Year', 'Sex'] + [col for col in df.columns if col not in ['Country', 'Year', 'Sex']]
df = df[new_order]

# make year an int type
df['Year'] = df['Year'].astype(int) 
# make country into category type
df['Country'] = df['Country'].astype("category")

df.head(5)
```

```python
# save dataset as cleaned
df.to_csv('data/clean_df.csv')
```

### Fixing column data types

```python
# make sex into 0 or 1
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = one_hot_encoder.fit_transform(df[['Sex']])

one_hot_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out(['Sex']))
df_onehot = pd.concat([df, one_hot_df], axis=1)

df_onehot
```

```python
# drop duplicate columns
df_onehot.drop(['Sex_Women', 'Sex'], axis = 1, inplace = True)
df_onehot['Sex'] = df_onehot['Sex_Men']
df_onehot.drop(['Sex_Men'], axis = 1, inplace = True)
df_onehot
```

In the column 'Sex', 0 represents female, 1 represents male

```python
df = df_onehot
# function to get the continent for each country code
def country_to_continent(country_code):
    try:
        # convert the 3-letter country code to country name
        country_name = pycountry.countries.get(alpha_3 = country_code)
        # convert country name to continent code
        continent_code = pc.country_alpha2_to_continent_code(country_name.alpha_2)
        return continent_code
    except:
        return None

# apply the function to the 'Country' column (assuming 3-letter country codes)
df['Continent'] = df['Country'].apply(country_to_continent)
df
```

```python
df.to_csv('data/df.csv')
```

```python
# group by country
df_drop = df.drop(columns=['Year', 'Sex', 'Continent'])
df_by_country = df_drop.groupby('Country').mean()
df_by_country.head(5)
df_by_country.reset_index(inplace=True)
df_by_country.head(5)
```

### Data Exploration

```python
# group by year and continent, calculate mean and std
continent_stats = df.groupby(['Year', 'Continent']).agg(
    mean_prevalence=('Crude diabetes prevalence', 'mean'),
    std_prevalence=('Crude diabetes prevalence', 'std')
).reset_index()

plt.figure(figsize=(10, 6))

# each continent's average prevalence over the years with shaded std
for continent in continent_stats['Continent'].unique():
    subset = continent_stats[continent_stats['Continent'] == continent]
    # plot mean
    plt.plot(subset['Year'], subset['mean_prevalence'], label=continent)
    # plot std as shaded fill
    plt.fill_between(
        subset['Year'], 
        subset['mean_prevalence'] - subset['std_prevalence'], 
        subset['mean_prevalence'] + subset['std_prevalence'], 
        alpha=0.2, 
        label='_nolegend_'
    )

plt.title('Crude Diabetes Prevalence Has Increased Over Time', fontsize = 16, fontname = 'monospace', fontweight = 'bold')
plt.xlabel('Year', fontsize = 14, fontname = 'monospace')
plt.xticks(fontname = 'monospace')
plt.ylabel('Average Crude Diabetes Prevalence', fontsize = 14, fontname = 'monospace')
plt.legend(labels=['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America'], prop={'family': 'monospace'}, loc='upper left')
plt.grid(True)
# save to figures folder
plt.savefig('figures/diabetes_increase.svg')
plt.show()
```

```python
# filter out data for the U.S. and U.K. 
df_us = df[df['Country'] == 'USA']
df_uk = df[df['Country'] == 'GBR']

# group the data by country and year
# calculate mean non-HDL cholesterol level
df_grouped = df.groupby(['Country', 'Year'])['Mean non-HDL cholesterol (mmol/L)'].mean().reset_index()

plt.figure(figsize=(11, 6))

# plot all countries in gray
for country in df_grouped['Country'].unique():
    df_country = df_grouped[df_grouped['Country'] == country]
    plt.plot(df_country['Year'], df_country['Mean non-HDL cholesterol (mmol/L)'], color='gray', alpha=0.3, linewidth=1)

# highlight U.S. in red
plt.plot(df_us['Year'], df_us['Mean non-HDL cholesterol (mmol/L)'], color='#ff005c', label='United States', linewidth=2.5)

# highlight U.K. in red
plt.plot(df_uk['Year'], df_uk['Mean non-HDL cholesterol (mmol/L)'], color='#016fff', label='United Kingdom', linewidth=2.5)

plt.title('Average Non-HDL Cholesterol Levels Decreasing Over Time in the U.S. and U.K.', fontsize=16, fontname = 'monospace', fontweight = 'bold')
plt.xlabel('Year', fontsize=14, fontname = 'monospace')
plt.xticks(fontname = 'monospace')
plt.ylabel('Average Non-HDL Cholesterol (mmol/L)', fontsize=14, fontname = 'monospace')
plt.legend(prop={'family': 'monospace'})
plt.grid(True)
plt.tight_layout()
plt.savefig('figures/cholesterol_levels.svg')
plt.show()
```

```python
sns.set_palette('colorblind')
# group the data by Continent and Sex
# calculate the mean for prevalence of hypertension
columns_of_interest = ['Prevalence of hypertension'] 
sex_stats = df.groupby(['Continent', 'Sex'])[columns_of_interest].mean().reset_index()

# pivot data
pivot_df = sex_stats.pivot(index='Continent', columns='Sex', values=columns_of_interest[0]).reset_index()

bar_width = 0.35
positions = range(len(pivot_df))
fig, ax = plt.subplots(figsize=(9, 6))

# plot bars for 'Men' (Sex 1 = men)
ax.bar(positions, pivot_df[1], width=bar_width, label='Male', color='#57b4e9')

# plot bars for 'Women' (Sex 0 = women)
# shift them by bar_width to the right
ax.bar([p + bar_width for p in positions], pivot_df[0], width=bar_width, label='Female', color='#fbafe4')

ax.set_xlabel('Continent', fontsize = 14, fontname = 'monospace')
ax.set_ylabel('Average Prevalence of Hypertension', fontsize = 14, fontname = 'monospace')
ax.set_title('Hypertension Prevalence is Highest in European Males', fontsize = 16, fontweight = 'bold', fontname = 'monospace')
ax.set_xticks([p + bar_width / 2 for p in positions])
ax.set_xticklabels(['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America'], fontname = 'monospace')
ax.set_axisbelow(True)
ax.legend(title='Sex', prop={'family': 'monospace'})
ax.grid(True)
plt.savefig('figures/hypertension_prevalence.svg')
plt.show()
```

```python
# examining hypertension prevalence sex differences (particularly in western diet continents EU and NA)

df_filtered = df[df['Continent'].isin(['EU', 'NA'])]

# group by Year, Continent, and Sex
# calculate both the mean and standard deviation of hypertension prevalence
sex_stats = df_filtered.groupby(['Year', 'Continent', 'Sex']).agg(
    mean_hypertension=('Prevalence of hypertension', 'mean'),
    std_hypertension=('Prevalence of hypertension', 'std')
).reset_index()

plt.figure(figsize=(10, 6))

# plot for each continent and sex
for continent in sex_stats['Continent'].unique():
    for sex in sex_stats['Sex'].unique():
        subset = sex_stats[(sex_stats['Continent'] == continent) & (sex_stats['Sex'] == sex)]
        
        # Plot the mean line
        plt.plot(subset['Year'], subset['mean_hypertension'], label=f"{continent} - {'Male' if sex == 1 else 'Female'}")
        
        # Add the translucent standard deviation as a shaded region
        plt.fill_between(
            subset['Year'], 
            subset['mean_hypertension'] - subset['std_hypertension'], 
            subset['mean_hypertension'] + subset['std_hypertension'], 
            alpha=0.3
        )

plt.title('Hypertension Prevalence by Sex in Europe and North America Over Time', fontsize = 16)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Mean Hypertension Prevalence', fontsize = 14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.set_palette('colorblind')

# group the data by continent and calculate the mean for interested columns
columns_of_interest = [
    'Prevalence of hypertension',
    'Proportion of diagnosed hypertension among all hypertension',
    'Proportion of treated hypertension among all hypertension',
    'Proportion of controlled hypertension among all hypertension',
    'Proportion of untreated stage 2 hypertension among all hypertension'
]

continent_stats = df.groupby('Continent')[columns_of_interest].mean().reset_index()

plt.figure(figsize=(12, 8))
bar_width = 0.15
positions = list(range(len(continent_stats)))

# define positions for the bars for each metric (5 metrics = 5 bar positions per group)
bar_positions = [positions]
for i in range(1, len(columns_of_interest)):
    bar_positions.append([x + bar_width * i for x in positions])

# plot each column as a set of bars with a specific position
for i, column in enumerate(columns_of_interest):
    plt.bar(bar_positions[i], continent_stats[column], width=bar_width, label=column)

plt.xlabel('Continent')
plt.ylabel('Mean Value')
plt.title('Comparison of Hypertension-Related Metrics per Continent')
plt.xticks([r + bar_width*2 for r in positions], continent_stats['Continent'])
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.set_palette('colorblind')

# group the data by continent and calculate the mean for interested columns
columns_of_interest = [
    'Prevalence of hypertension',
    'Proportion of diagnosed hypertension among all hypertension',
    'Proportion of treated hypertension among all hypertension',
    'Proportion of controlled hypertension among all hypertension',
    'Proportion of untreated stage 2 hypertension among all hypertension'
]

continent_stats = df.groupby('Continent')[columns_of_interest].mean().reset_index()

plt.figure(figsize=(12, 8))

# initialize bottom of the bars to zero (for stacking metrics)
bottom = None

for column in columns_of_interest:
    if bottom is None:
        bottom = continent_stats[column]  # first stack at the bottom
        plt.bar(continent_stats['Continent'], continent_stats[column], label=column)
    else:
        plt.bar(continent_stats['Continent'], continent_stats[column], bottom=bottom, label=column)
        bottom += continent_stats[column]  # add the next metric on top

plt.xlabel('Continent')
plt.ylabel('Mean Value')
plt.title('Stacked Bar Chart of Hypertension-Related Metrics by Continent')
plt.legend(loc='upper right', title='Metrics')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# define bins for Mean non-HDL cholesterol and Stage 2 hypertension
cholesterol_bins = pd.cut(df['Mean non-HDL cholesterol (mmol/L)'], bins=10)
hypertension_bins = pd.cut(df['Proportion of untreated stage 2 hypertension among all hypertension'], bins=10)

# create a frequency table/pivot table for binned data
frequency_table = pd.crosstab(cholesterol_bins, hypertension_bins)

plt.figure(figsize=(10, 8))
sns.heatmap(frequency_table, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap of Mean Non-HDL Cholesterol vs. Stage 2 Hypertension')
plt.xlabel('Proportion of Untreated Stage 2 Hypertension')
plt.ylabel('Mean Non-HDL Cholesterol')
plt.tight_layout()
plt.show()
```

```python
# create a new column 'Decade' by grouping years into decades
df['Decade'] = (df['Year'] // 10) * 10  # Groups by decade (e.g., 1990, 2000, 2010, etc.)

# list of decades to loop over
decades = df['Decade'].unique()

# set up a grid of subplots (e.g., 2x3 grid for multiple decades)
fig, axes = plt.subplots(1, 3, figsize=(11, 3))  # adjust grid size as needed
axes = axes.flatten()

colors = sns.color_palette('Set2', len(decades))

# loop over each decade
# plot the scatter plot in each subplot
for i, decade in enumerate(decades):
    # filter the data for the current decade
    df_decade = df[df['Decade'] == decade]
    
    # create scatter plot for each decade in its own subplot
    axes[i].scatter(
        df_decade['Mean non-HDL cholesterol (mmol/L)'], 
        df_decade['Proportion of untreated stage 2 hypertension among all hypertension'],
        color=colors[i],  # use a single color for all points in this decade
        s=50
    )
    
    # set title and labels for each subplot
    axes[i].set_title(f"{decade}s", fontsize=14)
    axes[i].set_xlabel('Mean Non-HDL Cholesterol (mmol/L)', fontsize=10)
    axes[i].set_ylabel('Proportion of Untreated Stage 2 Hypertension', fontsize=10)
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```
